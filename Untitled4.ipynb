```python
# Function to extract text from a pdf file
def extract_text(pdf_path):
    try:
        with open(pdf_path, 'rb') as file:
            reader = PyPDF2.PdfReader(file)
            text = ""
            for page in reader.pages:
                text +=page.extract_text()
            return text
    except Exception as e:
        return f"An error occured : {e}"

if __name__== "__main__": 
    pdf_path = "C:\\Users\\Poste\\Downloads\\diplome.pdf"
    text= extract_text(pdf_path)
    print (text)
    
```

```python
def outlier(data,column):
    mean = np.mean(data[column])
    std =  np.std(data[column])
    q1 = np.percentile(data[column],25)
    q3 = np.percentile(data[column],75)
    iqr = q3 - q1
    lowerbound = q1 - 1.5 * iqr
    upperbound = q3 + 1.5 * iqr
    outliers=[]
    outliers=data[(data[column]<lowerbound) | (data[column]>upperbound)]
    return outliers
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

data={'Departement' : ['Brazzaville', 'Pointe-Noire', 'Niari', 'Bouenza', 'Cuvette',
                    'Cuvette-Ouest', 'Plateaux', 'Likouala', 'Sangha', 'Pool'],
     'Pauvrete' : [15, 20, 45, 40, 60, 55, 65, 70, 62, 50],
     'Alphabetisation' : [95, 90, 65, 70, 50, 52, 45, 35, 40, 60]}

df=pd.DataFrame(data)
inertie=[]
n_samples = len(df)
max_clusters = min(9,n_samples)

for k in range(1,max_clusters + 1):
    model=KMeans(n_clusters=k)
    model.fit(df[['Pauvrete','Alphabetisation']])
    inertie.append(model.inertia_)

plt.plot(range(1,max_clusters+1), inertie, marker='o')
plt.xlabel('Number de clusters')
plt.ylabel('Inertia')
plt.title('Nombre de clusters')
```

```python
kmeans=KMeans(n_clusters=3,random_state=0)
df['Clusters']=kmeans.fit_predict(df[['Pauvrete','Alphabetisation']])
plt.figure(figsize=(8,6))
plt.scatter(df['Pauvrete'],df['Alphabetisation'], c=df['Clusters'], cmap='viridis')
for i in range(len(df)):
    plt.text(df['Pauvrete'][i]+0.5,df['Alphabetisation'][i]+0.5,df['Departement'][i], fontsize=9)
plt.xlabel('Taux de Pauvrete')
plt.ylabel('Taux d\'Alphabetisation')
plt.title('Pauvrete par Departement')
plt.grid("blue")
plt.show()
```
